```python
import pandas as pd
import numpy as np

# created a Sample DataFrame For Project
data = {
    'Date': pd.date_range('2023-01-01', periods=10),
    'Product': np.random.choice(['A', 'B', 'C'], size=10),
    'Quantity': np.random.randint(1, 10, size=10),
    'Revenue': np.random.uniform(50, 200, size=10)
}

df = pd.DataFrame(data)
df.to_csv('sales_data.csv', index=False)
```

```python
per75=np.percentile(df["Revenue"],75)
per25=np.percentile(df["Revenue"],25)
iqr=per75-per25
upper_limit=per75*1.5*iqr
print(upper_limit)
Revenue_per=df[df["Revenue"]>upper_limit]
Revenue_per
#to find  outlier in the Revenue
```

```python
#Box Plot to get a Clear Idea
import matplotlib.pyplot as plt
plt.boxplot(df["Revenue"])
plt.title(f"Revenue_chart from 23-01-01 to 23-01-10 ")
plt.xlabel('Revenue')
plt.show()
```

```python
#by looking this plot we can say from 23-01-01 to 23-01-10 Revenue lies between 50 to 190
```

```python
#bar plot to compare Quality and Revenue
Product_df=df.groupby("Product").sum().reset_index()
plt.plot(Product_df["Product"],Product_df["Quantity"])
```
