```python
import pandas as pd
import numpy as np

# created a Sample DataFrame For Project
data = {
    'Date': pd.date_range('2023-01-01', periods=10),
    'Product': np.random.choice(['Honda', 'Mahindra', 'Tata',"Mercides","Volswagn","pojero","Loamborgini","ferrari"], size=10),
    'Quantity': np.random.randint(1, 10, size=10),
    'Revenue': np.random.uniform(50, 200, size=10)
}

df = pd.DataFrame(data)
df.to_csv('sales_data.csv', index=False)
```

```python
per75=np.percentile(df["Revenue"],75)
per25=np.percentile(df["Revenue"],25)
iqr=per75-per25
upper_limit=per75*1.5*iqr
print(upper_limit)
Revenue_per=df[df["Revenue"]>upper_limit]
Revenue_per
#to find  outlier in the Revenue
```

```python
#Box Plot to get a Clear Idea
import matplotlib.pyplot as plt
plt.boxplot(df["Revenue"])
plt.title(f"Revenue_chart from 23-01-01 to 23-01-10 ")
plt.xlabel('Revenue')
plt.show()
```

```python
#by looking this plot we can say from 23-01-01 to 23-01-10 Revenue lies between 50 to 190
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df with columns like "Product" and "Quantity"
Product_df = df.groupby("Product")["Quantity"].sum().reset_index()

# Plotting the data
plt.bar(Product_df["Product"], Product_df["Quantity"],color="blue",linewidth=3)
plt.title(f" Product Average sales on Quantity",fontsize=20,color="green")
plt.xlabel('Product', fontsize=20,color="red")
plt.ylabel('Quantity', fontsize=20,color="red")
# Display the plot
plt.show()
```

```python
#in this graph as you can see the highest sales was done by Tata and the lowest sales was done by Pojero
```
